# Variable

We use variable in our program to remember some information. Our program can change the value of these variable dynamically. For example, if we make a game, and we want to keep track of how many lives a player has, we need to use variable to remember this information. 

Usually people explains variable as a storage box that you can store information in. This is half correct in Python. It is best to understand the variable as two part: name and value. 

* Value: That's the information your program keep in the storage box. Just like the items you keep in your Minecraft chest.
* Name: is NOT the storage box itself. Instead, it is the name we use to label the storage box, just like the sign we put on our Minecraft chest.

Then what about the storage box itself? Do we call that something? Well, that's the memroy in your computer, a piece of semiconductor IC chip.

## 2.4 String

String plays an important part in programming. Programs need to process string to extract information from external world, and it needs to format string in order to display information to human. 

Unlike numbers (like int and float) that is natual to computer, string are not. String are for human. Computer understand things in number. Therefore under the hood, computer convert string to number in order to tell what character it is. The conversion follows the ASCII table. 

![ascii table](https://cdn.shopify.com/s/files/1/1014/5789/files/Standard-ASCII-Table_large.jpg?10669400161723642407)

In this table, look at the **Dec** column and **Char** column. We see that a symbol ```$``` is represented by number 36. A letter ```e``` is represented by number 101.

Since computer recgonize everything by number, string must be categorize as a different datatype. 

* When Python know the datatype is string, and it sees number 36 in the memory, it converts that number to ```"$"```. 
* If Python knows the datatype is integer, and it sees number 36, it doesn't do any conversion.

Let's use Python code to checkout the ASCII table. Feel free to try different numbers from the ASCII table **Dec** column. 

````python
chr(56)
````

````python
a_number = 97
print("a_number is type",type(a_number))
print("Convert a_number to float:", float(a_number))
print("Convert a_number to string by ASCII table:", chr(a_number))
````

String is a series of character inside a pair of quotes. You can use single quote, double quotes, triple quotes. Python understand them all as string. 

````python
a_str = 'Hello world.'
b_str = "Nice to \n meet you."
c_str = """
My name is Henry.
I am from England
"""
#print(a_str, b_str, c_str)
print(a_str, '\n', b_str, c_str)
````

* This example shows three ways of creating a variable with string value. The triple quote allows us to write multi-line strings, while single quote and double quote can only be one line. 
* The second print statement used ```'\n'```. That is a special string telling Python to start a new line. Whenever Python sees '\n', it create a new line and continue printing.

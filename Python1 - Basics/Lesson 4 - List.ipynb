# Loops

Looping means doing some tasks again and again and again. 




One advantage of machine is they never get bored. They are excellent at doing repeated task. So instead of writing repeated code ourselves, let's get Python to repeat the code itself. This is done by using some looping keyword. 

##1.1 For Loop

When you know how many times you want to repeat, you can use ```for``` loop. 

````python
for i in range(10): # range(10) creates a sequence integer from 0 to 9
  print(i)
````

```For``` loop is most often used with multi-items sequence data type. When we don't have a pre-defined sequence, we use the range(4) to create a sequence of 4 numbers. When we have a pre-defined sequence, we can replace the range(4) with our sequence variable. 

````python
a_list = ["Good morning", 
          "Today is Friday", 
          "How are you?", 
          "Have a greate weekend!"] # create a list of 4 sentences.
````

## 1.4 Math with List and Loop

Many every day problem requires repeated math calculation, such as

- Calculating average grade of a class
- Calculating total expense on employee salary 
- Calculating total number of corn production of all the farm in a province

With loop and list, we can turn these problem into simple 3-4 lines of Python code. 



---

**Math Hint**

When we need to calculate the total of many number adding together, we call that **the sum**

```total = number1 + number2 + number3 + number4 +...```

If we need to calculate average, then we use the sum divided with the number of values used for calculating the sum

```
average = total / number of item
```
---

Now if we want to turn the total calculation into a loop, we need to turn the summing equation above into a recursive one. (*recursive means repeating*)

```
total_1 = 0 + number1
total_2 = total_1 + number2
total_3 = total_2 + number3
total_4 = total_3 + number4
and so on...
```
There you can see, to turn the summing equation into a loop, it is

```python
total = 0
for i in number_list:
  total = total + i
```

````python
num_list = [343,65,347,87,1,34,6,7,3,7,8,3,4,7,8,5,2,33]  
total=0                                                   
for num in num_list:
  print(f"old total is {total}")
  total = total+num                                  
  print(f"add number {num}, new total is {total}")
print(f"Result is {total}")
print(f"Average is {total/len(num_list)}")
````

## 1.2 While Loop

```while``` loop is used when we don't know how many times we want to repeat. Instead, the program decide when to quit looping using a conditional statement. 

```python
while <conditional>:
  # body of the loop
```

````python
a_num = 100
print("Enter a number bigger than 0 to keep looping")
while a_num>0:
  a_str = input("Enter a number:")
  a_num = int(a_str)
  print(f"You entered: {a_num}")
print(f"{a_num} is not bigger than 0, while loop exited.")
````

Practice writing a while loop which keep printing the altitude of a helicopter until the its altitude becomes 0, then print "We are landed". The psudo code looks like this:

```
Starting altitude is 100
```

````python
# Write your code here: 
import time
altitude = 100 # starting altitude for helicopter
# while loop simulating the helicopter landing
# comparison operator: <, >, ==, 
while altitude > 0:
  altitude = altitude - 5
  print(f"Current altitude is {altitude}")
  time.sleep(1)
print("We are landed")
````

# Dictionary

When we hear dictionary, we normally think of this: 

![dictionary](https://the1a.org/wp-content/uploads/sites/4/2019/05/GettyImages-630263206-768x525.jpg)

Each word in a dictionary has the 
```
{word:meaning}
```
In Python, there is also a data structure called dictionary. The dictionary has the form

```python
{key:value}
```

Dictionary is a Python data structure
* It can stores multiple items, enclosed by a pair of curly bracket
* Each item has a key and value separated by colon
* Key cannot be duplicated. Usually keys are strings. 
* Value can be any data type, ie. string, number, list, function, object, etc.
* Dictionary is changeable, expandable

Let's create a dictionary: 


````python
a_dict = {'Canada':'Toronto', 'Japan':'Tokyo', 'China':'Beijing', 'USA':'Washington D.C.', 'France':'Paris'}
````

````python
a_dict['Japan']
````

## Read, Modify, or Add to a Dictionary

Unlike list, dictionary has keys. Therefore each item in a dictionary can be read or modified by providing the key of that item. Just like the dictionary for language, you need to look up the word to get the meaning of the word. 

To read specific item in a dictionary, you would write

````python
a_dict['Korea']
````

To modify an item in a dictionary, you would write

````python
a_dict['Canada'] = 'Ottawa'
a_dict
````

To add a new item into a dictionary, you can write

````python
a_dict['Korea'] = 'Seoul'
a_dict
````

To delete an item from a dictionary, write

````python
del(a_dict['Canada'])
a_dict
````

* Key cannot be duplicated

````python
a_dict = {'Canada':'Ottawa', 'US':'Washington D.C.', 'Canada': 'Toronto'}
a_dict
````

# Loop through a Dictionary

To scan through each item in a dictionar, you can write a for loop this way: 

````python
for key, value in a_dict.items():
  print("Key is ", key, ", value is ", value)
````

## To Check if Certain Key Exists in a Dictionary

Sometimes we want to look for specific information in a dictionary.

For example, if the grades of the entire class is stored in a dictionary, and I want to look for the mark of student named Elizabeth. I can write the code this way:

```python
if key in a_dict:
  # do something with a_dict[key]
```


````python
marks = {'Vic': 'A', 'Addi':'B', 'Samuel':'A', 'Elizabeth':'B'}
if 'Elizabeth' in marks:
  print('Mark of Elizabeth is:', marks['Elizabeth'])
else:
  print('No such person')
# marks['Elizabeth']
````

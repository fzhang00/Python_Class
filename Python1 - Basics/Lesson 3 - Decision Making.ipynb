# How Program Make Decision

We can get program to make decision based on conditions. 

## 1. Conditional Statement

Conditional Statement are expression describing a condition. For example

* If we have a variable named ```temperature``` storing the latest temperature measurement. 

````python
temperature = 24
# This is a conditional statement checking for temperature bigger than 25
temperature > 25
````

Another example, 

* in a game, we have variable ```lives``` storing the number of lives a play has. Unfortunately the player just lost its last life

````python
lives = 1
# Another conditional statement checking for lives > 0
lives == 0
````

You can see the outcome of a conditional statement can only be two possibilities: 

* True: the variable satisfies the condition.
* False: the variable does not satisfy the condition

## 2. If Statement

Program uses conditional statement in if...else... statement to branch off to different tasks. 

The if statement is simple to understand. We use it in our daily decision making.

* **If** it is cloudy, I will bring a rain jacket when going outdoor; **otherwise (else)** I won't.
* **If** it is late, I'll go to bed; **otherwise (else)** I will play some video game. 

Program also use if statement in decision making. Let's look at how to write if statement in Python

```python
if <your_conditional_statement> :
  # perform some tasks here if
  # the condition is satisfied
else:
  # do some other tasks if
  # the condition is NOT satisfied
print("xyz")
```

## 3. Flow Chart

Flow chart helps us to visualize the flow of a program. 

```python

  # Do some tasks
else:
  # Do some other tasks 
```

We can also re-write the thermostats control into ```if...elif...else```

````python
temperature = input("Enter the current temperature:")
temperature = int(temperature)
if temperature > 25 :
  turn_on_ac()
  print("Air conditioner turned on.")
elif temperature < 24 : 
  turn_off_ac()
  print("Air conditioner turned off.")
elif temperature > 20 :
  turn_off_furnace()
  print("Furnace turned off.")
elif temperature < 16:
  turn_on_furnace()
  print("Furnace turned on.")
else:
  pass
print("end of program")
````

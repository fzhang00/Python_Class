# Debug

Debugging is an everyday practise of a programmer. In order to debug, you need to know how to read the traceback message. The traceback tells you what's wrong with your code. Traceback usually looks something like this: 

```python
  File "your_python_file_name", line 1
    print("Hello Sam"
                     ^
SyntaxError: unexpected EOF while parsing
```

- This first line tells you which file and on which line your error occured
- The second line prints out the actual code that causes the error
- The last line prints out the type of error, description and suggestion.

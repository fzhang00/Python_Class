# Function

We've learn the basic building blocks of Python language. Variable, conditional statement, ```if...else...``` statement, ```for``` loop, ```while``` loop are all fundermental building block of writing any program. They are like the brick and mortar when building a house. 

Let's introduce a new tool today. Function. Function is for organizing our code. They become very useful when our program grows long. 

Let's look at a example. 

# Case Study

Let's create a math quiz program that ask 10 summation questions, check player's answer and output scores. 

````python
from random import randint
n_questions = 10
score = 0
for i in range(n_questions):
  # Generate two random integer between 1 to 100 to create a question
  a = randint(1, 100)
  b = randint(1, 100)
  answer = input(f"{a} + {b} = ")
  # Check answer
  if int(answer) == (a+b):
    score = score + 1
  else:
    print("Uh oh!")
# Announce the result
print(f"You got {score} out of {n_questions} questions right! ")
````

We can see that in this small program, we did a number of specific tasks. We need to create a question, and we need to check player's answer.

To use built-in modules, we just need to import them directly.

# 3. How to use Modules?

In order to use module, we must first import them. When Python execute the line of importing module, it finds the module from your hard disk, run all the code of that module, and stores them in memory readily for your program to use. 

An example you have all encountered so far is the randint. Before you can use the randint(), you must first import it from the random module like this.

```python
from random import randint
```

Actually there are two ways to import modules. 

1. You can import the entire module. Then everything in this module, ie. all functions and constants, can be called by adding the module name with a ```.``` in front of the function, like this: 

```python
import random

print("random.randint output:", random.randint(0,100))
print("random.choice output:", random.choice(['1','b','12','Ottawa','capital']))
```

Run these code in the code cell below for multiple time to observe what they do. Pay attention that when I need to use function from the random module, I typed ```random.randint()``` and ```random.choice()```. This because I imported the entire random module by ```import random```

2. The second way you can import is to import specific items from the module, like this:

```python
from random import randint, choice

print("random.randint output:", randint(0,100))
```

In this case, when I need to use the function, I don't need to type ```random.``` anymore, just type the function name directly. 

However, only the function imported will be available. Other functions from the random module cannot be called.

3. Rename module during import

Sometimes the module name is long, or we use it very often in our code, that we don't want to type the full name every time. Imagine if you have to use functons from random module for a 100 times in your code...

In this case, we can rename the module during import like this:

```python
import random as rd
print("random.choice output:", rd.choice(['1','b','12','Ottawa','capital']))
```

Now the full module name ```random``` has been shrinked to just two letter ```rd```. Instead of calling ```random.randint()```, you would type ```rd.randint()```. Most people don't complaint when they just type two characters, and get to access all the functions in a module.
